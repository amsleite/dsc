##Tipos de problemas

###Diferenças entre grupos:
- Comparar se grupos são diferentes
- Exemplos de testes: teste T, ANOVA, teste de Mann-Whitney, teste exato de Fisher

###Classificação:
- Identificar a qual grupo o dado pertence
- Variável de saída assume valores discretos
- Exemplos de classificadores: naïve Bayes, k-nearest neighbours, SVM (support vector machine), regressão logística, classification tree (usa uma árvore de decisão), random forest classifier (usa várias árvores de decisão)
- Métricas: acurácia, precisão, sensitivity (recall), odds ratio, F1-score

###Regressão:
- Estimar ou prever resposta
- Variável de saída assume valores contínuos
- Exemplos de regressores: regressão linear, GLM (general linear model), ridge regression, modelos AR (auto-regressivos), ARMA (autoregressive moving average)
- Métricas: R-quadrado, R-quadrado ajustado, erro quadrático médio (MSE), erro médio absoluto (MAE)

###Aprendizado não-supervisionado:
- Não se conhecem as classes dos dados de entrada
- Exemplos: clusterização (k-means clustering), redução de dimensionalidade e blind signal separation (PCA, SVD, ICA)

##Machine learning
- Aprender algumas propriedades do conjunto de dados e aplicar em dados novos
- Para avaliar o algoritmo, os dados são separados em:
    - Conjunto de treinamento (~60%-80% dos dados): "aprender" as propriedades (e.g., encontrar os parâmetros do modelo)
    - Conjunto de teste: testar as propriedades "aprendidas"
- *Overfitting*: "aprender em excesso" com o conjunto de treinamento, ou seja, fazer um ajuste excessivo aos dados do treinamento dando menos margem para os dados de teste (dados novos)

```python
from sklearn import datasets # import standard datasets
from sklearn import svm # support vector machine
from sklearn import cluster # clustering
from sklearn.ensemble import RandomForestClassifier
import sklearn.metrics as metrics # scikit learn metrics
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
# Target names = classes
iris.target_names.tolist()
```

```python
# Feature names
iris.feature_names
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['class_no'] = iris.target
iris_df['class_name'] = np.nan
for ii, cc in enumerate(iris.target_names):
    iris_df.loc[iris_df['class_no']==ii, 'class_name'] = cc
#Processo inverso: pd.factorize(iris_df['class_name'])
#iris_df['class_name2'] = pd.Factor(iris.target, iris.target_names)
```

```python
g = sb.PairGrid(iris_df[iris.feature_names + ['class_name']], hue='class_name')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
# same results as: sb.pairplot(...)
```

```python
iris_df[iris.feature_names + ['class_name']].boxplot(by='class_name', figsize=(10,10));
```

```python
# Support vector machine
classif = svm.SVC()
X, y = iris.data, iris.target
classif.fit(X, y)
```

```python
classif.predict(X)
```

```python
print 'Score = %.2f%%' %(100*classif.score(X, y)) # mean accuracy = (no. hits)/(total)
# same as: 1 - np.sum(classif.predict(X) - y)*1./len(y)
```

Críticas:
- *score* pode ter muitos significados; é melhor escrever explicitamente o que foi medido
- Quando novos dados forem obtidos, qual será o desempenho do classificador? Será que "aprendemos" alguma coisa?

É preciso separar os dados em teste/treinamento!

```python
iris_df['is_train'] = np.random.RandomState(seed=666).uniform(low=0, high=1, size=len(iris_df)) <= .8 # 80% train, 20% test
train = iris_df[iris_df.is_train]
test  = iris_df[-iris_df.is_train]
```

```python
classif = svm.SVC()
classif.fit(train[iris.feature_names], train['class_name']) # ou "class_no"
print 'Mean accuracy = %.2f%% (train), %.2f%% (test)' %(100*classif.score(train[iris.feature_names], train['class_name']),\
                                                        100*classif.score(test[iris.feature_names], test['class_name']))
print 'Sample size = %d (train), %d (test)' %(len(train), len(test))
```

```python
iris_df.groupby(by='class_name').size().reset_index().rename(columns={0: 'Count'})
```

```python
train.groupby(by='class_name').size().reset_index().rename(columns={0: 'Count'})
```

```python
test.groupby(by='class_name').size().reset_index().rename(columns={0: 'Count'})
```

```python
y_pred = classif.predict(test[iris.feature_names])
print 'Mean accuracy =', metrics.accuracy_score(test['class_name'], y_pred)
print 'Mean F1 score =', metrics.f1_score(test['class_name'], y_pred)
```

```python
metrics.classification_report(test['class_name'], y_pred)
```

```python
# Matriz de confusão
metrics.confusion_matrix(test['class_name'], y_pred)
```

```python
pd.crosstab(test['class_name'], y_pred, rownames=['Medido'], colnames=['Previsto'])
```

```python
# Random forest classifier
rfc = RandomForestClassifier()
rfc.fit(train[iris.feature_names], train['class_name'])
y_pred_rfc = rfc.predict(test[iris.feature_names])
```

```python
pd.crosstab(test['class_name'], y_pred_rfc, rownames=['Medido'], colnames=['Previsto'])
```

```python
metrics.classification_report(test['class_name'], y_pred_rfc)
```

# Testes para comparar valores entre os grupos

```python
print 'T-test & Mann-Whitney test'

for ff in range(0, len(iris.feature_names), 1):
    print '\nFeature =', iris.feature_names[ff]
    for cc1 in range(0, len(iris.target_names), 1):
        for cc2 in range(cc1+1, len(iris.target_names), 1):
            print '%s vs. %s' %(iris.target_names[cc1], iris.target_names[cc2])
            print '    T, p_T = %.2f, %.2g' %stats.ttest_ind(iris_df.loc[iris_df['class_no']==cc1, iris.feature_names[ff]], \
                                            iris_df.loc[iris_df['class_no']==cc2, iris.feature_names[ff]], equal_var=False)
            print '    U, p_U = %.2f, %.2g' %stats.mannwhitneyu(iris_df.loc[iris_df['class_no']==cc1, iris.feature_names[ff]], \
                                            iris_df.loc[iris_df['class_no']==cc2, iris.feature_names[ff]])
```

```python
print 'One-way ANOVA and Kruskal-Wallis H-test'
# Two-way ANOVA: examines the influence of 2 different categorical independent variables on 1 continuous dependent variable 
# MANOVA (multivariate analysis of variance): compares multivariate (population) means of several groups.
# As a multivariate procedure, it is used when there are 2 or more dependent variables, although statistical reports provide
# individual p-values for each dependent variable in order to test for statistical significance.
print 'Comparing all groups:'
print iris.target_names

for ff in range(0, len(iris.feature_names), 1):
    print '\nFeature =', iris.feature_names[ff]
    print '    F, p_F = %.2f, %.2g' %stats.f_oneway(iris_df.loc[iris_df['class_no']==0, iris.feature_names[ff]],\
                                                   iris_df.loc[iris_df['class_no']==1, iris.feature_names[ff]],\
                                                   iris_df.loc[iris_df['class_no']==2, iris.feature_names[ff]])
    print '    H, p_H = %.2f, %.2g' %stats.kruskal(iris_df.loc[iris_df['class_no']==0, iris.feature_names[ff]],\
                                                   iris_df.loc[iris_df['class_no']==1, iris.feature_names[ff]],\
                                                   iris_df.loc[iris_df['class_no']==2, iris.feature_names[ff]])
```
